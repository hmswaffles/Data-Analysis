```python
from __future__ import division
import prettyplotlib as ppl
import numpy as np
import scipy.stats as sps
import time
import csv

# prettyplotlib imports
import matplotlib.pyplot as plt
import matplotlib as mpl
from prettyplotlib import brewer2mpl
%matplotlib inline

def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
```

```python
#Getting the data
d = open("C:\Users\Evan\Desktop\B2BTimes.csv",'r')
dd = csv.reader(d)
mdata = []
fdata = []
#splitting into male and female times
for row in dd:
    mdata.append(row[0:3])
    if row[4]:
        fdata.append([row[0],row[3],row[5]])


```

```python
def convert_times(d):
    ###converts times to seconds
    records = []

    for row in d:
        t = row[2]
        if len(t) <5:
            print "Hello"
        else:
            if t[0] == '1':
                (h,m,s)=t.split(':')
                seconds = int(h)*3600+int(m)*60+int(s)
            elif t[5] == ':':
                (m,s,ts)=t.split(':')
                seconds = int(m)*60+int(s)
            elif t[5] == '.':
                (m,s)=t.split(':')
                seconds = int(m)*60+float(s)
            else:
                print t
            records.append(seconds)
    return records

```

```python
def convert_dates(d):
    dates = []
    #print len(d)
    P=len(d)-14
    for i in range(len(d)):
       # print i
        
        row = d[i]
        date = row[0]
        year = date[-2:]
        #print year
        if i >P-2:
            full_year = '20'+year
        else:
            full_year = '19'+year
        dates.append(full_year)
    return dates
```

```python
def correct_times(gender,d):
    speedls = []
    if gender=='m':
        for i in range(len(d)):
            if i<71:
                speed = float(d[i])/7.51
            else:
                speed = float(d[i])/7.46
            speedls.append(speed)
    elif gender == 'w':
        for i in range(len(d)):
            if i<15:
                 speed = float(d[i])/7.51
            else:
                speed = float(d[i])/7.46
            speedls.append(speed)
            
    return speedls
```

```python
def make_plots(d,correct):
    fig, ax = plt.subplots(1)
    men = d[0]
    women = d[1]
    if correct:
        
        ay = convert_times(men)
        Ym = correct_times('m',ay)
        Xm = convert_dates(men)
        print "hola"
        print len(Xm)
        print len(Ym)
        ppl.scatter(ax,Xm,Ym,label="Men's speeds")

        aw = convert_times(women)
        Yw = correct_times('w',aw)
        Xw = convert_dates(women)

        ppl.scatter(ax,Xw,Yw,label="Women's speeds")

        ppl.legend(ax)
        ax.set_title("Bay To Breakers Speeds (Seconds per Mile)")
        fig.savefig("b2bwinningtimescorrected.png")

        

    else:
        Ym = convert_times(men)
        Xm = convert_dates(men)
        ppl.scatter(ax,Xm,Ym,label="Men's Times")

        Yw = convert_times(women)
        Xw = convert_dates(women)
        ppl.scatter(ax,Xw,Yw,label="Women's Times")

        ppl.legend(ax)
        ax.set_title("Bay To Breakers times (seconds)")
        fig.savefig("b2bwinningtimes.png")

   #fig.show()

    
```

```python
mspeeds = Ym
spearmanls = []
spearmanp = []
a = range(1,11)
for j in range(len(mspeeds)-10):
    speed_window = mspeeds[j:j+10]
    len(speed_window)
    corr,pvalue = sps.spearmanr(speed_window,a)
    spearmanls.append(corr)
    spearmanp.append(pvalue)
```

```python
setup_graph(title="Spearman's Rho, buckets of 10",fig_size=(12,8))

ppl.scatter(range(1903,1903+len(spearmanls)),spearmanls,label = 'Coefficent')
```

```python
spearman_meta_ls = []
spearman_meta_p = []
for k in range(10,40,4):
    a = range(1,k+1)
    s_ls = []
    s_p = []
    for j in range(len(mspeeds)-k) :
        speed_window = mspeeds[j:j+k]
        corr,pvalue = sps.spearmanr(speed_window,a)
        s_ls.append(corr)
        s_p.append(pvalue)
    spearman_meta_ls.append(s_ls)
    spearman_meta_p.append(s_p)
```

```python
cmapp = [[250,250,199],[255,237,160],[254,217,118],[254,178,76],[253,141,60],[252,78,42],[227,26,28],[189,0,38],[128,0,38]]
ncmap =[]
for color in cmapp:
    newcolor = map(lambda x: x/255.0,color)
    ncmap.append(newcolor)
    

  
```

```python
setup_graph(title="Spearman's Rho, Variable buckets",fig_size=(12,8))

for i in range(len(spearman_meta_ls)):
    current= spearman_meta_ls[i]
    ppl.plot(range(1903,1903+len(current)),current,label=i,color=ncmap[i+1])
    plt.legend(loc='best')
len(spearman_meta_ls)
```

```python
setup_graph(title="Evenly Starting Rho, Variable buckets",fig_size=(12,8))

for i in range(len(spearman_meta_ls)):
    current= spearman_meta_ls[i]
    currentlen = len(current)
    size = 103 - currentlen
    half = size/2
    x = range(1903+int(half),1903+103-int(half))    
    ppl.plot(x,current,label=i,color=ncmap[i+1])
    plt.legend(loc='best')
```

```python
#### What about significance? Let's scale the correlations by their pvalue.
setup_graph(title="Rho*P, evenly spread, swept-through buckets",fig_size=(12,8))

for i in range(len(spearman_meta_ls)):
    current= spearman_meta_ls[i]
    currentp = spearman_meta_p[i]
    c = [a*(1-b) for a,b in zip(current,currentp)]
    currentlen = len(current)
    size = 103 - currentlen
    half = size/2
    x = range(1903+int(half),1903+103-int(half))    
    ppl.plot(x,c,label=i,color=ncmap[i+1])
    plt.legend(loc='best')
```
